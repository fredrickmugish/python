```python
import numpy as np
speed  = [99,86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86 ]
x = np.mean(speed)
print(x)
```

```python
import numpy as np
from scipy import stats
speed  = [99,86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86 ]
y = stats.mode(speed)
x = np.std(speed)
z = np.var(speed)
print("The standard deviation is: " +str(x))
print("The mode is: " +str(y))
print("The variance is: " +str(z))
```
